```python
import requests
root_url = "https://api.binance.com/"
```

```python
# Getting Response from Exchange Info API

exchangeInfo_endpoint = "api/v1/exchangeInfo"
exchangeInfo_response = requests.get(root_url+exchangeInfo_endpoint)
exchangeInfo_response = exchangeInfo_response.json()
print(len(exchangeInfo_response['symbols']))
# Generating a set for all available symbols

symbols  = set()
createstrings = []
for coin in exchangeInfo_response['symbols']:
    symbols.add(coin['baseAsset'])
    symbols.add(coin['quoteAsset'])
    createstrings.append(coin['symbol'])

no_of_symbols = len(symbols)
# print(no_of_symbols)

# print(symbols)
print(createstrings)
```

```python
# Mapping Symbols to Numbers in symbol_map dictionary

symbol_map = {}
counter = 0

for symbol in symbols:
    symbol_map[symbol] = counter
    counter+=1
    
# print(symbol_map)
  
# Initialising Adjacency Matrix
price_matrix = [[float('inf')]*no_of_symbols for symbol in range(no_of_symbols)]
```

```python
ticker_price_endpoint = "api/v3/ticker/price"
ticker_price_response = requests.get(root_url+ticker_price_endpoint).json()
print(len(ticker_price_response))
l = []
baseAssets = set()
quoteAssets = set()
for index in range(len(ticker_price_response)) :
    baseAsset = exchangeInfo_response['symbols'][index]['baseAsset']
    baseAssets.add(baseAsset)
    quoteAsset = exchangeInfo_response['symbols'][index]['quoteAsset']
    quoteAssets.add(quoteAsset)
    price_matrix[symbol_map[baseAsset]][symbol_map[quoteAsset]] = ticker_price_response[index]['price']
```

```python
# Create an Adjacency list, quoteAsset:{baseAsset1,baseAsset2}
price_list = {}
for index in range(len(ticker_price_response)) :
    baseAsset = exchangeInfo_response['symbols'][index]['baseAsset']
    quoteAsset = exchangeInfo_response['symbols'][index]['quoteAsset']
    if quoteAsset in price_list:
        price_list[quoteAsset].append(baseAsset)
    else:
        price_list[quoteAsset] = [baseAsset]
print(len(price_list))
```

```python
# Implementing Dijkstra

mat =  [[ 0, 4, 0, 0, 0, 0, 0, 8, 0 ],
        [ 4, 0, 8, 0, 0, 0, 0, 11, 0 ],
        [ 0, 8, 0, 7, 0, 4, 0, 0, 2 ],
        [ 0, 0, 7, 0, 9, 14, 0, 0, 0 ],
        [ 0, 0, 0, 9, 0, 10, 0, 0, 0 ],
        [ 0, 0, 4, 14, 10, 0, 2, 0, 0 ],
        [ 0, 0, 0, 0, 0, 2, 0, 1, 6 ],
        [ 8, 11, 0, 0, 0, 0, 1, 0, 7 ],
        [ 0, 0, 2, 0, 0, 0, 6, 7, 0 ]];

def dijkstra(source, mat):
    visited = set()
    visited.add(source)
    dist = [float('inf') for idx in range(len(mat))]
    dist[source] = 0
    while len(visited)<len(mat):
        print(len(visited))
        min_dist_node = 0
        min_dist = float('inf')
        for node in range(len(dist)):
            if dist[node]<min_dist and node not in visited:
                min_dist = dist[node]
                min_dist_node = node
        for node in range(len(mat[min_dist_node])):
            if dist[min_dist_node]+mat[min_dist_node][node]< dist[node] and mat[min_dist_node][node]!=0:
                dist[node] = dist[min_dist_node]+mat[min_dist_node][node]
        visited.add(min_dist_node)

    print(dist)
dijkstra(0,mat)
            
```
